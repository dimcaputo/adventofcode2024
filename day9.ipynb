```python
import os
import sys
```

```python
with open(os.path.expanduser('~/Downloads/input9.txt')) as f:
    txt = ''
    for x in f.read():
        txt += x
```

```python
sys.set_int_max_str_digits(20000)
```

```python
def make_long_string(string):   
    new_string = ''
    ind = 0
    for i in range(len(string)):
        if string[i] != '\n' and i%2 == 0:
            new_string += str(ind)*int(string[i])
            ind += 1
        if string[i] != '\n' and i%2 != 0:
            new_string += '.'*int(string[i])
    return new_string
```

```python
long_string = make_long_string(txt)
```

```python
def move_chars(string):
    template = list(string)
    j = -1
    i = 0
    new_string = ''
    while i != j and i <= len(template) - 1 and j >= -len(template): 
        if template[i] == '.' and template[j] != '.':
            template[i],template[j]  = template[j], template[i]
            i += 1
            j -= 1
        elif template[i] == '.' and template[j] == '.':
            j -= 1
        elif template[i] != '.' and template[j] != '.':
            i += 1
        elif template[i] != '.' and template[j] == '.':
            i += 1
            j -= 1
        else:
            print('Nope')
    for item in template:
        new_string += item
    return new_string

```

```python
condensed = move_chars(long_string)
```

```python
def sum_chars(string):
    sum_prod = 0
    for i in range(len(string)):
        if string[i] != '.':
            sum_prod += i*int(string[i])
    return sum_prod
```

```python
sum_to_find = sum_chars(condensed)
```
